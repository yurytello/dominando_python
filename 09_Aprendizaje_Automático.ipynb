## Regresión Lineal

En estadística, la regresión lineal o ajuste lineal es un modelo matemático usado para aproximar la relación de dependencia entre una variable dependiente Y, con variables independientes X.

Este método es aplicable en muchas situaciones en las que se estudia la relación entre dos o más variables o predecir un comportamiento

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Crear un conjunto de datos sencillo (por ejemplo, años y sus ingresos asociados)
#     X representa los años (1, 2, 3, ...)
X = np.array([1, 2, 3, 4]).reshape(-1, 1)  # Reshape porque scikit-learn requiere 2D array

#     y representa los resultados correspondientes (por ejemplo, unidades monetarias)
y = np.array([0.20, 0.40, 0.60, 0.80])

# 2. Crear el modelo de regresión lineal
modelo = LinearRegression()

# 3. Ajustar (fit) el modelo a los datos
modelo.fit(X, y)

# 4. Hacer una predicción para un nuevo valor (por ejemplo, predecir para el año 5)
prediccion = modelo.predict([[6]])

# Imprimir la predicción
print(f"Predicción para el año 6: {prediccion[0]} unidades monetarias")
```

```python
 # Visualizar los datos y la línea de regresión
 plt.scatter(X, y, color='blue', label='Datos reales')  # Gráfica de puntos
 plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión')  # Línea de regresión

 for anho in range(5,7):
     prediccion = modelo.predict([[anho]])
     plt.scatter(anho, prediccion, color='green', label=f'Predicción para año {anho}: {prediccion[0]}')

 plt.title("Regresión Lineal Sencilla")
 plt.xlabel("Años")
 plt.ylabel("Unidades Monetarias")
 plt.legend()
 plt.show()
```

# scikit-learn

### 1. Obtenemos el Conjunto de Datos

```python
import numpy as np

X = np.random.random((10,5))
y = np.array(['F','F','F','M','M','M','M','M','M','M'])
print(y)

print(f"X original:\n{X}")
X[X < 0.7] = 0
print(f"X modificado:\n{X}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

print(f"X_train:\n{X_train}")
print(f"y_train:\n{y_train}")
print(f"X_test:\n{X_test}")
print(f"y_test:\n{y_test}")
```

### 2. Seleccionamos el Modelo

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

3. Ajustamos el modelo a los datos

```python
#print(f"X_train:\n{X_train}")
#print(f"y_train:\n{y_train}")
svc.fit(X_train, y_train)

y_train_numerico = np.where(y_train == 'M', 1, 0)
#print(f"y_train_numerico:\n{y_train_numerico}")

lr.fit(X_train, y_train_numerico)
```

### 4. Hacemos las Predicciones

```python
X_random = np.random.random((3,5))
X_random[X_random < 0.7] = 0
print(f"X_random:\n{X_random}")

y_pred = svc.predict(X_random)
print(f"\nSVC \ny_pred:\n{y_pred}")

y_pred = lr.predict(X_random)
print(f"\nRegreson Lineal \ny_pred:\n{y_pred}")

y_pred = lr.predict(X_test)
print(f"\nRegreson Lineal (test) \nX_test:\n{X_test}\ny_pred:\n{y_pred}")
```

### 5. Evalualua el Modelo

```python
from sklearn.metrics import mean_absolute_error

# Transforma a numérico el y_test
y_test = np.where(y_test == 'M', 1, 0)

error = mean_absolute_error(y_test, y_pred)

print(f"Error Absoluto: {round(error * 100)}% ")
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)

print(f"Coeficiencite de Determinación R^2: {round(score * 100)}%")
```
